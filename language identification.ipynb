```python
import pandas as pd
import nltk
```

```python
df = pd.read_csv("dataset (1).csv")
df
```

```python
df1 = df.drop_duplicates()
```

```python
df1
```

```python
x = df1['Text']
x
```

```python
y = df1[['language']]
y
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
LB = LabelEncoder()
y1 = LB.fit_transform(y)
y1
```

```python
y2 = pd.DataFrame(y1,columns=['language'])
y2.value_counts()
```

# tokenization

```python
from nltk.tokenize import sent_tokenize
```

```python
def Tokenization(data): 
    tokens = sent_tokenize(data)
    return tokens
```

```python
x_token = x.apply(Tokenization) #take x as series
x_token
```

# cleaning

```python
from string import punctuation
```

```python
def Cleaning(data):
    clean = [i for i in data if i not in  punctuation]
    return clean
```

```python
x_clean = x_token.apply(Cleaning)
x_clean
```

```python
def Join(data):
    text = ' '.join(data)
    return text
```

```python
x_join = x_clean.apply(Join)
x_join
```

# splittg the data

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x_join,y2,test_size=0.2,random_state=42,stratify=y2)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
TF = TfidfVectorizer(max_df=0.95,max_features=1000)
```

```python
xtraintf = TF.fit_transform(xtrain)
xtesttf = TF.transform(xtest)
```

```python
TF.get_feature_names()
```

```python
xtf = pd.DataFrame(xtraintf.A,columns=TF.get_feature_names())
xtf
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
mnb_model = MultinomialNB()
mnb_model.fit(xtraintf.A,ytrain)
```

```python
from sklearn.metrics import accuracy_score
```

```python
ypredtest = mnb_model.predict(xtesttf.A)
```

```python
accuracy_score(ytest,ypredtest)
```

# Finding solution when input is given

```python
sent = input()
def solution(model,sentence):
    token = sent_tokenize(sentence)
    clean = [i for i in token if i not in  punctuation]
    text = ' '.join(clean)
    text = TF.transform([text])
    y = model.predict(text)
    return LB.inverse_transform(y)


print('The selected language is',solution(mnb_model,sent))
```
